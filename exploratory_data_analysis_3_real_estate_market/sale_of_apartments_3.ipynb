## Предобработка данных <a id='2-bullet'></a>

```python
df.duplicated().sum() #проверка дубликатов
df['balcony'] = df['balcony'].fillna(0).astype('int') # пропуск заменен на ноль, изменение типа данных со строчного на временной.
df['days_exposition'] = df['days_exposition'].fillna(0).astype('int')

df = df.drop(columns=['is_apartment', 'parks_around3000', 'parks_nearest', 'ponds_around3000', 'ponds_nearest', 'airports_nearest'])
# не будут использованы в текущем исследовании

# обработка пропусков в высоте потолков
df.ceiling_height.unique() # отмечено, что есть двузначные и очень маленькие цифры, скорее всего опечатки
df.loc[df['ceiling_height'].isnull(),'ceiling_height']= df['ceiling_height'].median() #без разбивки на группы, тк независимо от числе этажей

#Расстояние до центра: медианным по городу заполнить нельзя, тк есть населенные пункты без данных. 
#Для анализа нужны только данные по Санкт-Петербургу, определено, что таких 3% - можно удалить нулевые строки без потери
round(len(df[(df['locality_name'] == 'Санкт-Петербург') & (df['cityCenters_nearest'].isna())]) / len(df.reset_index(drop=True).query('locality_name == "Санкт-Петербург"')), 3)


#Данные важные, но пропуски нельзя обработать, от города и локации зависит стоимость.
# Общее число этажей в некоторой степени коррелирует с этажом, но замена по граппам приведет к тому, что нижние этажи войдут не в те группы, поэтому пропущенные строки удаляю как не информативные
df.dropna(subset = ['locality_name', 'cityCenters_nearest', 'floors_total'], inplace=True)

#рассмотрим медианное значение жилой площади в зависимости от числа комнат, найдем коэффициент соотношения
print ((df.loc[(df['rooms'] == 0) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 0) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 1) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 1) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 2) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 2) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 3) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 3) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 4) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 4) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 5) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 5) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 6) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 6) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 7) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 7) & (df['living_area'].isnull() == False), 'total_area']).median())
print ((df.loc[(df['rooms'] == 9) & (df['living_area'].isnull() == False), 'living_area']/df.loc[(df['rooms'] == 9) & (df['living_area'].isnull() == False), 'total_area']).median())


#заполнение пропусков жилой площади медианным значением
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 0), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 0), 'total_area'] * 0.69
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 1), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 1), 'total_area'] * 0.48
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 2), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 2), 'total_area'] * 0.59
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 3), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 3), 'total_area'] * 0.63
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 4), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 4), 'total_area'] * 0.66
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 5), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 5), 'total_area'] * 0.65
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 6), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 6), 'total_area'] * 0.68
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 7), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 7), 'total_area'] * 0.68
df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 9), 'living_area'] = df.loc[(df['living_area'].isnull() == True) & (df['rooms'] == 9), 'total_area'] * 0.63

#статистика размеров квартир для группировки по квантилям
print(df['total_area'].describe())

#выбираем медианные по группам для соотношения
print((df.loc[(df['total_area'] <= 40), 'kitchen_area']/df.loc[(df['total_area'] <= 40), 'total_area']).median())
print((df.loc[(df['total_area'] > 40) & (df['total_area'] <= 52), 'kitchen_area']/df.loc[(df['total_area'] > 40) & (df['total_area'] <= 52), 'total_area']).median())
print((df.loc[(df['total_area'] > 52) & (df['total_area'] <= 70), 'kitchen_area']/df.loc[(df['total_area'] <= 70) & (df['total_area'] > 52), 'total_area']).median())
print((df.loc[(df['total_area'] > 70) & (df['total_area'] <= 150), 'kitchen_area']/df.loc[(df['total_area'] > 70) & (df['total_area'] <= 150), 'total_area']).median())
print((df.loc[(df['total_area'] > 150), 'kitchen_area']/df.loc[(df['total_area'] > 150), 'total_area']).median())

#заполнение пропусков медианным по группам, умноженным на соотношение
df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] <= 40), 'kitchen_area'] = df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] <= 40), 'total_area'] * 0.24
df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] >40) & (df['total_area'] <= 52), 'kitchen_area'] = df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] >40) & (df['total_area'] <= 52),'total_area'] * 0.165
df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] >52) & (df['total_area'] <= 70), 'kitchen_area'] = df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] >52) & (df['total_area'] <= 70),'total_area'] * 0.15
df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] >70) & (df['total_area'] <= 150), 'kitchen_area'] = df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] >70) & (df['total_area'] <= 150),'total_area'] * 0.14
df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] >150), 'kitchen_area'] = df.loc[(df['kitchen_area'].isnull() == True) & (df['total_area'] > 150), 'total_area'] * 0.11




df.info()
df.isnull().sum()
```

### Вывод

1.  `Дубликаты не обнаружены`
2. `Пропуски в числе балконов и числе дней размещения` — заменены на ноль.  
Неуказанное число балконов может быть заменено на ноль, т.к. пропуск как правило означает их отсутствие.  
Пропуск дней размещения может быть в том случае, если оно менее одного дня.  
Этот показатель фиксируется автоматически и может быть вызван либо реальным снятием квартиры с продажи, либо техническим сбоем. 
Рассматривается именно снятие объявления (после нахождения покупателя) его автором, а не юридическая продажа квартиры, которая не может быть в пределах одного дня.
Основная масса публикующих объявления о продаже квартир, как правило, является профессиональными посредниками (риэлторами), что нужно принимать во внимание при анализе данных.
Такое может быть довольно часто, сразу в ряде случаев. К какому именно варианту относится отдельный пропуск достоверно определить незможно, но можно выделить причины возможной быстрой продажи квартиры:  
* во-первых, размещение квартиры происходит сразу на нескольких сервисах и во внутренних базах риелторских компаний, но не одновременно, поэтому часть квартир может быть дольше в продаже на других сайтах. 
* во-вторых,размещение в сервисе Яндекс-Недвижимость дает возможность получение скидок и льгот у партнеров сервиса, в том числе и на ипотечное кредитование, поэтому квартира может найти покупателя на момент, а размещаться для получения скидок
* в-третьих, на рынке действует крупный игрок срочного выкупа недвижимости (ПИК-брокер), а также ряд мелких компаний по срочному выкупу квартир по рыночной цене. И доля рынка ПИК-брокера растет. Как правило, часть клиентов стремятся продать жилье как можно быстрее, поэтому соглашаются на продажу, по оценкам агентств потери в цене составляют всего 9,5–10%.  
Согласно данным Циан, такими услугами очень часто пользуются собственники квартир с "проблемной историей".
* в четвертых, рост спроса на жилье приводит к быстрому "вымыванию" наиболее выгодных вариантов. Например, при открытии новых веток метро, присоединении областных объектов к юридической черте города и пр.  
* мошенничество. Размещение несуществующих объектов недвижимости для изменения цен на рынке.
Принимая во внимание растущий рынок срочного выкупа замена на медианное значение не проводилась. Данные о мошенничестве как правило выявляются при проверке объявлений, но в данном случае не указано, была ли проверка и снято ли обьявление сами пользователем или модератером сайта.  
В дальнейшем, данные менее 1 дня были отфильтрованы.
3. `Пропуски в высоте потолков` — заменены на медианное значение.
Анализ уникальных значений высоты потолков показал, что значительной разницы от медианного значения нет, кроме того, высота потолков не зависит напрямую от числа этажей или других факторов. Для анализа факторов, влияющих на стоимость квартиры построим матрицу корреляции цены жилья и квадратного метра от площади, комнат, удаленности от центра. Для анализа полученных значений (силы связи) будет использована шкала Чеддока.  
Основные гипотезы корреляционного анализа цены квартиры:
* между ценой квартиры и ее площадью будет прямая связь: чем ни большей площадь квартиры, тем она дороже стоит.
* между ценой квартиры и число комнат в ней будет прямая связь: чем ни большей площадь квартиры, тем она дороже стоит.
* между ценой квартиры и удаленностью от центра будет обратная связь: чем ни дальше квартира от центра, тем она дешевле.  

Рассмотрим корреляцию факторов и для цены квадратного метра, описательно, не строя гипотез, так как в данных не содержится разделение на типы недвижимости и можно предположить искажение результатов.

```python
display(df.loc[:,['last_price', 'total_area','rooms','cityCenters_nearest']].corr().round(2))
df.loc[:,['price_per_meter', 'total_area','rooms','cityCenters_nearest']].corr().round(2)
```

На полную стоимость квартиры ожидаемо наибольшее прямое влияние оказывает общая площадь (прямая высокая связь коэффициент корреляции 0,8): чем ни больше метраж, тем дороже квартира. Аналогично увеличивает стоимость число комнат (прямая заметная связь, коэффициент 0,51). Чем ни дальше квартира находится от центра города, тем она дешевле (обратная умеренная связь -0,35). Указанные факторы действуют на цену как новостроек, так и вторичек.  
В тоже время, если анализировать цену квадратного метра, то указанная зависимость будет искажена отсутствием данных о типе жилья (вторичка или новостройка): по отдельности в каждой из группы могла получить подтверждения гипотеза об обратной связи между стоимость квадратного метра и площадью. А именно, чем ни больше площадь, тем меньше стоимость за квадратный метр, соответственно существующим экономическим реалиям. Но смешавшись между собой без возможности разделения, вторичная и первичная недвижимость разного класса, а также находящейся в разных локациях не будет зависмой от метража, что и подтвердилось данными корреляции (отсутствие связи, коэффициент 0,05 < 0,1, минимального значения для подтверждения гипотезы о наличии связи между факторами).  
Проверка гипотезы зависимости цены за квадратный метр в пределах одной локации на примере центра Санкт-Петербурга будет проведена в конце исследования.  
В тоже время данная гипотеза косвенно подверждается наличием слабой отрицательной связи между числом комнат и ценой за квадратный метр. Связь между удаленностью от центра и ценой квадратного метра выражена еще ярче, она обратная и умеренная. При увеличении отдаленности от центра, стоимость квадратного метра падает.

Основные гипотезы:

Исходя из граика, выделим локации не более 9 км от центра, примерно в радиусе до конца Адмиралтейского района, находящегося в центре.

```python
center_spb_data = df.query('cityCenters_nearest_km <= 9 and locality_name == "Санкт-Петербург"')
plot_my_hist(center_spb_data['total_area'])
plot_my_hist(center_spb_data['rooms'])
plot_my_hist(center_spb_data['last_price'])
plot_my_hist(center_spb_data['ceiling_height'])
```
